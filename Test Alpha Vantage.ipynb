# Read daily adjusted closing prices

```python
import requests

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=IBM&outputsize=full&apikey=SXC3G8LJRFJT20W5'
r = requests.get(url)
data = r.json()

print(data.keys())
# print()
# print('*' * 50)
# print(data['Time Series (Daily)'])
# print()

```

```python
daily_ibm = pd.DataFrame.from_dict(data['Time Series (Daily)']).T.astype(float)
daily_ibm.sort_index(inplace=True)
display(daily_ibm.describe())
display(daily_ibm.head())


daily_ibm['5. adjusted close'].loc['2015-01-01':].plot(
    figsize=(15, 6),
    grid=True, 
    title='Raw data closing prices.'
);
plt.show();


daily_ibm['5. adjusted close'].loc['2015-01-01':].rolling(30).mean().plot(
    figsize=(15, 6), 
    grid=True, 
    title='Moving average.'
);
plt.show();


daily_ibm['5. adjusted close'].loc['2015-01-01':].rolling(30).std().plot(
    figsize=(15, 6),
    grid=True, title='Moving std.');
plt.show();


ret_close = daily_ibm['5. adjusted close'].pct_change()
ret_close['moving_avg'] = ret_close.to_numpy().rolling(30).mean()
# ret_close['moving_std'] = ret_close.rolling(30).std()


ret_close.plot(title='Returns.', figsize=(15, 6), subplots=True);
plt.show();
```
